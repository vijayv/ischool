```python
import numpy as np
import pandas as pd
import datetime as dt
%pylab inline
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cat scrobbles.tsv | cut -f1,2,3,5,7,9,11,13,15 > subset_scrobbles.tsv
scrobbles = pd.read_csv("data/subset_scrobbles.tsv", delimiter="\t")
amy_scrobbles = pd.read_csv("amy/data/subset_scrobbles.tsv", delimiter="\t")
scrobbles["ISO time"] = pd.to_datetime(scrobbles["ISO time"])
amy_scrobbles["ISO time"] = pd.to_datetime(amy_scrobbles["ISO time"])
amy_scrobbles.head()
```

```python
amy_scrobbles.dtypes
```

## By Year

```python
scrobbles["year"] = scrobbles["ISO time"].dt.year
scrobbles["month"] = scrobbles["ISO time"].dt.month
scrobbles["plays"] = 1

amy_scrobbles["year"] = amy_scrobbles["ISO time"].dt.year
amy_scrobbles["month"] = amy_scrobbles["ISO time"].dt.month
amy_scrobbles["plays"] = 1
```

```python
by_artist_year = pd.pivot_table(scrobbles, index="artist name", cols="year", values="plays", aggfunc="sum").fillna(0)
amy_by_artist_year = pd.pivot_table(amy_scrobbles, index="artist name", cols="year", values="plays", aggfunc="sum").fillna(0)
```
